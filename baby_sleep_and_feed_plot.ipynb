```python
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import Legend, PrintfTickFormatter
from bokeh.models.tickers import FixedTicker
```

```python
data = pd.read_csv('sleep-eat.csv', index_col='time')
sleepx = []
sleepy = []
feedx = []
feedy = []
for i, row in data.iterrows():
    for j, value in row.iteritems():
        if value == 'S' or value == 's':
            sleepy.append(i)
            sleepx.append(float(j)/30)
        elif value == 'E' or value == 'e':
            feedy.append(i)
            feedx.append(float(j)/30)
```

```python
output_notebook()
```

```python
colors = ['#2D8E65', '#8E2D56']
p1 = figure(title="Baby Sleep and Feed", plot_width=950, plot_height=350,
           x_axis_label='Month of Life', y_axis_label='Time of Day',
           x_range=(2.75, 18), y_range=(24,0))
sleep = p1.rect(sleepx, sleepy, 1/30, .25, color=colors[0], width_units="data", height_units="data",
        muted_color=colors[0], muted_alpha=0.2, line_alpha=0)
feed = p1.rect(feedx, feedy, 1/30, .25, color=colors[1], width_units="data", height_units="data",
        muted_color=colors[1], muted_alpha=0.2, line_alpha=0)
legend = Legend(items=[("Sleep", [sleep]), ("Feed", [feed])], orientation="horizontal", click_policy="mute")
p1.add_layout(legend, 'below')
p1.yaxis[0].formatter = PrintfTickFormatter(format="%2f:00")
p1.yaxis.ticker = FixedTicker(ticks=[0, 6, 12, 18, 24])
show(p1)
```
